```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrape_bbc_news(url):
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        articles = soup.find_all('div', class_='gs-c-promo')

        titles = []
        contents = []
        topics = []

        for article in articles:
            title_element = article.find('h3', class_='gs-c-promo-heading__title')
            title = title_element.text.strip() if title_element else None

            content_element = article.find('p', class_='gs-c-promo-summary')
            content = content_element.text.strip() if content_element else None

            topic_element = article.find('a', class_='gs-c-section-link')
            topic = topic_element.text.strip() if topic_element else None

            if title and content and topic:
                titles.append(title)
                contents.append(content)
                topics.append(topic)

        df = pd.DataFrame({'Title': titles, 'Content': contents, 'Topic': topics})
        return df
    else:
        print(f"Failed to fetch data from {url}. Status Code: {response.status_code}")
        return None

url = 'https://www.bbc.com/news'
bbc_news_data = scrape_bbc_news(url)

# Save the data to a CSV file
if bbc_news_data is not None:
    bbc_news_data.to_csv('bbc_news_data.csv', index=False)
    print("Data saved successfully as 'bbc_news_data.csv'")
else:
    print("Failed to save data. Please check the scraping function.")
```

```python
import pandas as pd

# List of file names
file_names = ['bbc_news_data.csv', 'bbc_news_data1.csv', 'bbc_news_data2.csv', 'bbc_news_data3.csv', 'bbc_news_data4.csv']

# Initialize an empty list to store dataframes
dfs = []

# Iterate over file names and read CSV files into dataframes
for file_name in file_names:
    df = pd.read_csv(file_name)
    dfs.append(df)

# Concatenate all dataframes
merged_df = pd.concat(dfs, ignore_index=True)

# Save the merged dataframe to a new CSV file
merged_df.to_csv('news_article.csv', index=False)

print("Merged data saved successfully as 'news_article.csv'")
```
